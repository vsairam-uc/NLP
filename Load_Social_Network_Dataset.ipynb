```python
import pandas as pd
import requests
from io import StringIO
```

```python
def load_dataset():
  url = 'https://query.data.world/s/kz5ie7mk3vgbt5qxfvu3y5infmkv3f?dws=00000'
  # Download the CSV file
  response = requests.get(url)
  csv_data = StringIO(response.text)

  # Read the CSV file into a pandas DataFrame
  return pd.read_csv(csv_data)
```

```python
df = load_dataset()
```

```python
# Example: Count the number of tweets for each sentiment
sentiment_counts = df['airline_sentiment'].value_counts()
print("\nSentiment distribution:")
print(sentiment_counts)
```

```python
# Example: Get the top 5 reasons for negative sentiments
negative_reasons = df[df['airline_sentiment'] == 'negative']['negativereason'].value_counts().head()
print("\nTop 5 reasons for negative sentiments:")
print(negative_reasons)
```
